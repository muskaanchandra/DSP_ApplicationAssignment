```python
from PyEMD import EMD
import numpy  as np
import pylab as plt
```

```python
IMF = np.loadtxt(open("/NASDAQ_imf.csv", 'rb'), delimiter=',', skiprows=0)
```

```python
IMF = np.transpose(IMF)
```

```python
data = pd.read_csv ("/NASDAQ.csv")
```

```python
t = data.iloc[:,0].values
s = data.iloc[:,1].values
print(t)
print(s)
```

```python
N = IMF.shape[0]+1
```

```python
# Plot results
f, axs = plt.subplots(N,N,figsize=(15,30))
plt.subplot(N,1,1)
plt.plot(t, s, 'r')
plt.title("Input signal: Closing prices from NASDAQ index")
plt.xlabel("Time [s]")

for n, imf in enumerate(IMF):
    plt.subplot(N,1,n+2)
    plt.plot(t, imf, 'g')
    plt.title("IMF "+str(n+1))
    plt.xlabel("Time [s]")

plt.tight_layout()
plt.show()
```
